```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
import numpy as np
```

## IMPORT DATA

```python
def train_model(data: pd.DataFrame = None, 
                fill_strategy: str = None,
                model: str = None):
    """Trains a model on the given data and returns performance metrics.
    
    Parameters:
    data (pd.DataFrame): The data to use for training.
    fill_strategy (str): The strategy to use for filling missing values.
    model (str): The name of the model.
    
    Returns:
    dict: A dictionary containing the model name, fill strategy, and performance metrics.
    """
    
    # Fill missing values in the F21 column using the specified strategy
    if fill_strategy=='mean':
        data['F21'] = df['F21'].fillna(data['F21'].mean())
    elif fill_strategy=='mode':
        data['F21'] = df['F21'].fillna(data['F21'].mode()[0])
    elif fill_strategy=='median':
        data['F21'] = df['F21'].fillna(data['F21'].median())
    elif fill_strategy=='remove':
        data = data.drop('F21', axis=1)
    
    # Split the data into features and labels
    features = data.drop('Class', axis=1)
    labels = data['Class']

    # Split the data into training and test sets
    X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size=0.1, random_state=428969)

    # Create a decision tree classifier
    dtc = DecisionTreeClassifier(random_state=42, max_depth=500)

    # Fit the classifier to the training data
    dtc.fit(X_train, y_train)

    # Make predictions on the test data
    pred = dtc.predict(X_test)
    
    # Make predictions on the training data
    pred_train = dtc.predict(X_train)
    
    # Calculate performance metrics for the training data
    train_accuracy = accuracy_score(y_train, pred_train)
    
    # Calculate performance metrics for the test data
    accuracy = accuracy_score(y_test, pred)

    # Return the model name, fill strategy, and performance metrics
    return {'Model': model,
            'Strategy': 'None' if fill_strategy == 'remove' else fill_strategy, 
            'Train Accuracy': train_accuracy,
            'Accuracy': accuracy}
```

```python
without_column = train_model(model='DecisionTreeClassifier',
                             data=df,
                             fill_strategy='remove'
                             )

with_mean = train_model(model='DecisionTreeClassifier',
                             data=df,
                             fill_strategy='mean')

with_median = train_model(model='DecisionTreeClassifier',
                             data=df,
                             fill_strategy='median')

with_mode = train_model(model='DecisionTreeClassifier',
                             data=df,
                             fill_strategy='mode')
```

```python
def train_more_models(data: pd.DataFrame = df,
                      test_size: int = None
                      ):
    """Trains multiple models on the given data and returns their performance scores.
    
    Parameters:
    data (pd.DataFrame): The data to use for training.
    
    Returns:
    pd.DataFrame: A dataframe containing the performance scores for each model.
    """
    
    # Define a dictionary of models to train
    model_list = {'decision-tree': DecisionTreeClassifier(), 
                  'logistic-regression': LogisticRegression(max_iter=1000, solver='liblinear'), 
                  'gradient-boosting': GradientBoostingClassifier(subsample=0.5, n_estimators=1000), 
                  'random-forest': RandomForestClassifier(),
                  'ada-boost': AdaBoostClassifier(),
                  'svc': SVC(kernel='linear')}
    
    # Split the data into features and labels
    features = data.drop(['F21', 'Class'], axis=1)
    labels = data['Class']

    # Split the data into training and test sets
    X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size=test_size, random_state=42)
    
    # Initialize a dictionary to store the model results
    model_results = {}
    
    # Train each model and store their performance scores
    for model_name, model in model_list.items():
        model.fit(X_train, y_train)
        
        # Make predictions on the training and test data
        pred_train = model.predict(X_train)
        pred = model.predict(X_test)
        
        # Calculate the accuracy scores for the training and test data
        accuracy_train = accuracy_score(pred_train, y_train)
        accuracy_test = accuracy_score(pred, y_test)
        
        # Add the accuracy scores to the model results dictionary
        model_results.update({model_name: [accuracy_train, accuracy_test]})
    
    # Convert the model results to a dataframe and rename the rows
    model_results = pd.DataFrame(model_results)
    model_results = model_results.rename(index={0: 'train_score', 1: 'test_score'})
        
    return model_results
```

```python
# Training with 80% of the samples
models_80 = train_more_models(test_size=0.2)
print('Sample size = 80%')
models_80
```

```python
# Training with 90% of the samples
models_01 = train_more_models(test_size=0.1)
print('Sample size = 90%')
models_01
```
